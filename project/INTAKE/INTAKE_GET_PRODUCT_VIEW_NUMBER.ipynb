```python
from selenium import webdriver
import gc
import pickle
import time
```

```python
options = webdriver.ChromeOptions()
options.add_argument('headless')
options.add_argument('window-size=1920x1080')
options.add_argument("disable-gpu")
```

```python
product_view = dict()
url = 'https://www.shopintake.com/product/view/'

driver = webdriver.Chrome("C:/chromedriver.exe")#, chrome_options=options)

for n in (i for i in range(1,1000)):
    if n % 50 == 0:
        print(n)
        gc.collect()
    try:
        driver.get(url+str(n)+'/')
        try:
            driver.switch_to_alert().accept()
        except:
            pass
        product = driver.find_element_by_xpath('//*[@id="contents"]/div[2]/div/div[2]/h2')
        product_title = product.text
        product_view[n] = product_title
    except:
        continue
        
driver.close()
```

```python
with open('pv_number.pickle', 'wb') as f:
    pickle.dump(product_view, f)
```

```python
from bs4 import BeautifulSoup as bs

event_view = dict()
url = 'https://www.shopintake.com/event/?&page='
driver = webdriver.Chrome('C:/chromedriver.exe')
for i in range(1,6):
    driver.get(url+str(i))
    page = driver.page_source
    HTML = bs(page, 'html.parser')
    target = HTML.find('div', {'class':'event_list'})
    target = target.findAll('div',{'class':'textarea'})
    for j in target:
        event_view[j.find('a')['href'].split('/')[3]] = j.find('a').get_text().strip() 
    time.sleep(1)
driver.close()
```
